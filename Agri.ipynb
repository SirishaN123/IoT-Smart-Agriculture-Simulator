```python
!pip install pandas matplotlib --quiet

# Import dependencies
import random
import pandas as pd
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output, display

# -------------------------------
# Step 1: Generate IoT Sensor Data
# -------------------------------
def generate_sensor_data(n=50):
    """Simulate IoT sensor data for smart agriculture"""
    data = {
        "Temperature (°C)": [],
        "Soil Moisture (%)": [],
        "Humidity (%)": [],
        "Light (Lumen)": []
    }
    for _ in range(n):
        data["Temperature (°C)"].append(round(random.uniform(15, 40), 1))   # Range 15-40°C
        data["Soil Moisture (%)"].append(round(random.uniform(20, 80), 1)) # Range 20-80%
        data["Humidity (%)"].append(round(random.uniform(30, 90), 1))      # Range 30-90%
        data["Light (Lumen)"].append(random.randint(100, 1000))            # Brightness
    return pd.DataFrame(data)

sensor_data = generate_sensor_data(50)

# -------------------------------
# Step 2: Simulate Real-Time Data Streaming
# -------------------------------
print("📡 Streaming IoT sensor data...\n")
for i in range(1, 6):
    clear_output(wait=True)
    display(sensor_data.head(i*10))  # Show data in batches of 10
    time.sleep(1)

# -------------------------------
# Step 3: Visualize IoT Data
# -------------------------------
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0,0].plot(sensor_data["Temperature (°C)"], color='red')
axs[0,0].set_title("Temperature (°C)")

axs[0,1].plot(sensor_data["Soil Moisture (%)"], color='brown')
axs[0,1].set_title("Soil Moisture (%)")

axs[1,0].plot(sensor_data["Humidity (%)"], color='blue')
axs[1,0].set_title("Humidity (%)")

axs[1,1].plot(sensor_data["Light (Lumen)"], color='orange')
axs[1,1].set_title("Light Intensity (Lumen)")

plt.tight_layout()
plt.show()

# -------------------------------
# Step 4: Smart Alert System
# -------------------------------
alerts = []
for i, row in sensor_data.iterrows():
    if row["Temperature (°C)"] > 35:
        alerts.append(f"[Row {i}] 🔴 High Temp Alert: {row['Temperature (°C)']}°C")
    if row["Soil Moisture (%)"] < 30:
        alerts.append(f"[Row {i}] 🟠 Low Soil Moisture Alert: {row['Soil Moisture (%)']}%")
    if row["Humidity (%)"] < 40:
        alerts.append(f"[Row {i}] 🔵 Low Humidity Alert: {row['Humidity (%)']}%")
    if row["Light (Lumen)"] < 200:
        alerts.append(f"[Row {i}] 🟢 Low Light Alert: {row['Light (Lumen)']} Lumen")

print("\n📢 Alerts Generated:\n")
for alert in alerts:
    print(alert)
```
